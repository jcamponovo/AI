### Affichage de la table ASCII
Les 32 premiers caractères n'apparaissent pas mais ont d'autres fonctions comme le retour à la ligne (Line Feed) par exemple.
Ils sont tous codés sur un octet, c'est à dire une série de 8 bits.

```python
table = ["NUL\t\t","SOH\t\t","STX\t\t","ETX\t\t","EOT\t\t","ENQ\t\t","ACK\t\t","BEL\t\t","Backspace\t","Horizontal Tab\t","Line Feed\t","Vertical Tab\t","Form Feed\t","Carriage Return","Shift Out\t","Shift In\t","DLE\t\t","DC1\t\t","DC2\t\t","DC3\t\t","DC4\t\t","NAK\t\t","SYN\t\t","ETB\t\t","CAN\t\t","EM\t\t","SUB\t\t","ESC\t\t","FS\t\t","GS\t\t","RS\t\t","US\t\t","Space\t\t"]+[chr(i)+"\t\t" for i in range(33,127)]+["DEL\t\t"]
print("Code\t","Caractère\t","Code Binaire")
print("")
for i in range(128):
    print("{:03d}\t".format(i),table[i],"{:08d}".format(int(bin(i)[2:])))
```





### Taille d'un fichier texte
Tapez un texte ci-dessous pour visualiser la taille en mémoire du fichier et le code binaire correspondant

```python
lin =     Textarea(
    value="A",
    placeholder='Ecrire un texte',
    description='Texte',
    disabled=False,
    layout=Layout(width='95%', height='100px')
)
lout =     Textarea(
    value="01000001",
    placeholder=' ',
    description='Code binaire',
    disabled=True,
    layout=Layout(width='95%', height='100px')
)
lout2 =     Textarea(
    value="1 octet",
    placeholder=' ',
    description='Taille fichier',
    disabled=True,
    layout=Layout(width='95%', height='50px')
)
lout3 =     Textarea(
    value="A",
    placeholder=' ',
    description='Texte décodé',
    disabled=True,
    layout=Layout(width='95%', height='100px')
)

encodage = ToggleButtons(
    options=['ASCII', 'ISO-8859-15', 'UTF-8'],
    description='Encodage:',
    disabled=False,
    button_style='',
)

decodage = ToggleButtons(
    options=['ASCII', 'ISO-8859-15', 'UTF-8'],
    description='Decodage:',
    disabled=False,
    button_style='',
)

def onclick(b):
    global codage
    if encodage.value=="ASCII":
        codage = "ASCII"
    elif encodage.value=="ISO-8859-15":
        codage = "iso8859_15"
    else:
        codage = "utf8"
    on_val_change(None)
    
def onclick2(b):
    global codage2
    if decodage.value=="ASCII":
        codage2 = "ASCII"
    elif decodage.value=="ISO-8859-15":
        codage2 = "iso8859_15"
    else:
        codage2 = "utf8"
    on_val_change(None)
    
    
encodage.observe(onclick,names='value')
decodage.observe(onclick2, names='value')

def on_val_change(change):
    global codage, codage2
    liste = list(lin.value)
    for i in range(len(liste)):
        if liste[i]=="\n":
            liste[i]="\r".encode(codage)+"\n".encode(codage)
        else:
            try:
                liste[i]=liste[i].encode(codage)
            except:
                liste = []
                lout.value = "IMPOSSIBLE"
                lout2.value = "IMPOSSIBLE"
                lout3.value = "IMPOSSIBLE"
                return
    liste2 = b''
    for val in liste:
        liste2 = liste2 + val
    listeval = [bin(val)[2:] for val in liste2]
    try:
        listeval2 = [val.decode(codage2) for val in liste]
    except:
        listeval2 = ["IMPOSSIBLE"]

    for i in range(len(listeval)):
        while len(listeval[i])<8:
            listeval[i]= "0"+listeval[i] #zfill(8)
    lout.value=" ".join(listeval)
    lout2.value="{:d} octet".format(len(listeval))+"s"*(len(listeval)>1)
    lout3.value="".join(listeval2)
    
codage = "ASCII"
codage2 = "ASCII"
display(encodage)
display(lin,lout,lout2)
display(decodage)
display(lout3)
lin.observe(on_val_change,names='value')
```
