### Affichage de la table ASCII
Les 32 premiers caractères n'apparaissent pas mais ont d'autres fonctions comme le retour à la ligne (Line Feed) par exemple.
Ils sont tous codés sur un octet, c'est à dire une série de 8 bits.

```python
table = ["NUL\t\t","SOH\t\t","STX\t\t","ETX\t\t","EOT\t\t","ENQ\t\t","ACK\t\t","BEL\t\t","Backspace\t","Horizontal Tab\t","Line Feed\t","Vertical Tab\t","Form Feed\t","Carriage Return","Shift Out\t","Shift In\t","DLE\t\t","DC1\t\t","DC2\t\t","DC3\t\t","DC4\t\t","NAK\t\t","SYN\t\t","ETB\t\t","CAN\t\t","EM\t\t","SUB\t\t","ESC\t\t","FS\t\t","GS\t\t","RS\t\t","US\t\t","Space\t\t"]+[chr(i)+"\t\t" for i in range(33,127)]+["DEL\t\t"]
print("Code\t","Caractère\t","Code Binaire")
print("")
for i in range(128):
    print("{:03d}\t".format(i),table[i],"{:08d}".format(int(bin(i)[2:])))
```





### Taille d'un fichier texte
Tapez un texte ci-dessous pour visualiser la taille en mémoire du fichier et le code binaire correspondant

```python
lin =     Textarea(
    value="A",
    placeholder='Ecrire un texte',
    description='Texte',
    disabled=False,
    layout=Layout(width='95%', height='100px')
)
lout =     Textarea(
    value="01000001",
    placeholder=' ',
    description='Code binaire',
    disabled=True,
    layout=Layout(width='95%', height='100px')
)
lout2 =     Textarea(
    value="1 octet",
    placeholder=' ',
    description='Taille fichier',
    disabled=True,
    layout=Layout(width='95%', height='50px')
)

def on_val_change(change):
    liste = list(lin.value)
    for i in range(len(liste)):
        if liste[i]=="\n":
            liste[i]="\r\n"
    liste = list("".join(liste))
    listeval = [ord(val) for val in liste]
    liste2 = []
    for val in listeval:
        if val<256:
            liste2.append("{:08d}".format(int(bin(val)[2:])))
        elif val<65536:
            liste2.append("{:016d}".format(int(bin(val)[2:])))
    lout.value=" ".join(liste2)
    lout2.value="{:d} octet".format(len("".join(liste2))//8)+"s"*(len("".join(liste2))//8>1)
display(lin,lout,lout2)
lin.observe(on_val_change,names='value')
```
